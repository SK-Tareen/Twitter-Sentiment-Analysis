### Feature Extraction with DistilBERT

```python
model_ckpt = "distilbert-base-uncased" # download distilbert tokenizer since they have their own
tokenizer = AutoTokenizer.from_pretrained(model_ckpt) # creation of the tokenizer
```

Now, the 'tokenize' method transforms each example in the dataset into a format suitable for input into the DistilBERT model. This transformation involves tokenizing the text, which means converting the raw text into numerical representations (tokens) that the model can understand.

- padding=True --> ensures that all sequences in the batch have the same length by adding padding tokens to shorter sequences.
- truncation=True --> specifies that if a sequence exceeds the model's maximum input length, it should be truncated (shortened) to fit within that limit.
- return_tensors="pt" --> indicates that the output should be returned as PyTorch tensors.

```python
def tokenize(batch):
    return tokenizer(batch["text"], padding=True, truncation=True, return_tensors="pt")
```

```python
# ensure that the dataset is prepared to work with PyTorch tensors:
dataset_dict.set_format(type="torch")

# applies the tokenize function to the dataset_dict to transform each example in the dataset (to tokens_ids)
tweets_encoded = dataset_dict.map(tokenize, batched=True, batch_size=None)
```

```python
print(tweets_encoded['train'].column_names) # Now, we have the dataset adapted to be used
# as input for DistilBERT.
```

Once we have the entire dataset adapted to be used as input for a DistilBERT, we will insert it into the complete DistilBERT model (this time we will not only use its tokenizer) so that it returns the representation of each of the sentences we use.

```python
import torch
import torch.nn.functional as F
from transformers import AutoModel

#load the entire model of DistilBERT
model_ckpt = "distilbert-base-uncased"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AutoModel.from_pretrained(model_ckpt).to(device)
```

- torch.no_grad() --> with this method PyTorch does not track operations on tensors, which can significantly reduce memory usage and speed up computations.

- batch --> expected to be a dictionary containing the input data for the model in a tokenized format.

- [CLS]('classification token') --> typically added at the beginning of every input sequence, and acts as a summary representation of the input sequence.

```python
def extract_hidden_states(batch):
    # Place model inputs on the GPU
    inputs = {k:v.to(device) for k,v in batch.items() # places the model inputs into de GPU
              if k in tokenizer.model_input_names}
    # Extract last hidden states
    with torch.no_grad():
        last_hidden_state = model(**inputs).last_hidden_state # feeds the inputs into the model
    # Return vector for [CLS] token
    return {"hidden_state": last_hidden_state[:,0].cpu().numpy()} # extracts the hidden state corresponding to the [CLS] token, which is located at index 0 of the second dimension of the last_hidden_state tensor
```

- set_format --> we are configuring the dataset to be compatible with PyTorch, allowing it to be used efficiently in model training.

- input_ids: it's the numerical representation of the tokenized input text.

- attention_mask: a tensor that indicates which tokens should be attended to (1 for actual tokens and 0 for padding tokens).

- sentiment: represents the target labels or categories for classification tasks, indicating the true labels associated with each tweet.

It extraccts the hidden states from a model for each input example in the dataset.
